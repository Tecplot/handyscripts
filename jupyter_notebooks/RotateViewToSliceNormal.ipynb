```python
import math
import tecplot as tp
tp.session.connect()
plot = tp.active_frame().plot()
slice = plot.slice(0)
x_norm, y_norm, z_norm = slice.arbitrary_normal

plot.view.psi = math.degrees(math.acos(z_norm))
plot.view.theta = -90-math.degrees(math.atan2(y_norm,x_norm))
plot.view.alpha = 0
plot.view.fit()
```
